```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
dataframe = pd.read_csv("test_data_MKGO.csv", sep = ';')
```

```python
dataframe = dataframe[dataframe.DateTime != "DateTime"]
dataframe = dataframe[dataframe.iloc[:, 1] != "Data Source"]
```

```python
dataframe = dataframe.drop_duplicates(subset=['DateTime', 'Data Source'])
```

```python
dataframe
```

```python
dataframe.loc[dataframe['Value'] == 'ON'] = 1
dataframe.loc[dataframe['Value'] == 'OFF'] = 0

# dataframe.at[dataframe['Value'] == 'ON', 'Value'] = 1 
# dataframe.at[dataframe['Value'] == 'OFF', 'Value'] = 0
```

```python
dataframe = dataframe[dataframe['Data Source'].notna()]
```

```python
points = dataframe['Data Source'].unique()
```

```python
points
```

```python
%%time
point_key = []
for i in range(len(points)):
    point_key.append('Point_' + str(i+1))
```

```python
point_dic = dict(zip(point_key, points))
```

```python
point_dic
```

```python
%%time
dataframe['DateTime'] = pd.to_datetime(dataframe.DateTime)
```

```python
%%time

dataframe['Value'] = dataframe['Value'].astype(float)
```

```python
datetimes = dataframe["DateTime"].unique()
```

```python
datetimes
```

```python
datetimes_sorted = pd.to_datetime(datetimes, format="%Y%m%d %H:%M:%S").sort_values()
```

```python
datetimes_sorted
```

```python
df = pd.DataFrame()
```

```python
df
```

```python
%%time
df['DateTime'] = datetimes_sorted
```

```python
df
```

```python
for key, point in point_dic.items():
    try:
        point_df = dataframe.loc[dataframe['Data Source'] == point][['DateTime', 'Value']]
        point_df.rename(columns = {'Value':key}, inplace = True)
        df = pd.merge(df, point_df, how='left', left_on='DateTime', right_on='DateTime', validate = 'm:1')
        
        print(key)
    except:
        pass
```

```python
point_df
```

```python
df = df.sort_values("DateTime")
```

```python
df
```

```python
df = df.fillna(0)
```

```python
df
```

```python
df['<>Date'] = pd.to_datetime(df['DateTime']).dt.date
df['Time'] = pd.to_datetime(df['DateTime']).dt.time
```

```python
df
```

```python
df.insert(0, 'Time', df.pop('Time'))
```

```python
df.insert(0, '<>Date', df.pop('<>Date'))
```

```python
df = df.drop(columns='DateTime')
```

```python
df = df.drop_duplicates(subset=['<>Date', 'Time'], keep='last')
```

```python
point_info_list = list(point_dic.items())
```

```python
point_info_df = pd.DataFrame(point_info_list)
```

```python
point_info_df
```

```python
first_row_df = pd.DataFrame(['Key            Name:Suffix                                Trend Definitions Used'])
```

```python
last_row_df = pd.DataFrame([' ******************************** End of Report *********************************'])
```

```python
pd.concat([point_info_df, df], axis = 0).to_csv('asdlkfj.csv', index = False, header=True)
```

```python
writer = pd.ExcelWriter('out.xlsx', engine='xlsxwriter')
```

```python
first_row_df.to_excel(writer, sheet_name='Sheet1', index=False, header=False)
point_info_df.to_excel(writer, sheet_name='Sheet1', startrow=1, index=False, header=False)
df.to_excel(writer, sheet_name='Sheet1', startrow=len(points)+5, index=False, header=True)
last_row_df.to_excel(writer, sheet_name='Sheet1', startrow=len(points)+5+len(datetimes)+1, index=False, header=False)
```

```python
writer.save()
```
